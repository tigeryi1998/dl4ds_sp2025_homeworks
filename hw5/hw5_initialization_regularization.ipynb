# **5.1 Weight Initialization**

```python
import numpy as np
import matplotlib.pyplot as plt

# ------- ACTIVATION FUNCTIONS -------
def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)
```

```python
# ------- INITIALIZATION FUNCTIONS -------

def init_random(fan_in, fan_out):
    """
    'Random' init: small Gaussian, scale=0.01
    Just for comparison (not specifically Xavier or He).
    """
    return np.random.randn(fan_in, fan_out) * 0.01
```

## 5.1.1 Complete the init_xavier function for Xavier Initialization

See Section 7.5 in UDL. Then look at the
initialization section of Chapter 7 Notes (at the end of the chapter).

```python
def init_xavier(fan_in, fan_out):
    """
    Xavier (Glorot) normal initialization
    """

    sigma_sq_omega = 1 / fan_in

    sigma_omega = np.sqrt(sigma_sq_omega)

    weights = np.random.normal(0, sigma_omega, (fan_in, fan_out))

    # weights = np.random.randn(fan_in, fan_out) * np.sqrt(1 / fan_in)

    return weights
```

See Section 7.5, 7.5.1 and in particular equation 7.32 in UDL.


```python
def init_he(fan_in, fan_out):

    """
    He (Kaiming) normal initialization
    """

    sigma_sq_omega = 2 / fan_in

    sigma_omega = np.sqrt(sigma_sq_omega)

    weights = np.random.normal(0, sigma_omega, (fan_in, fan_out))

    # weights = np.random.randn(fan_in, fan_out) * np.sqrt(2 / fan_in)

    return weights
```

```python
# ------- FORWARD-PASS HELPER -------
def forward_pass(dims, init_func, activation_func):
    """
    Given a list of layer dimensions [d0, d1, d2, ...],
    repeatedly initialize W for each layer with init_func,
    apply x = activation_func(x.dot(W)), track stats.

    Returns:
      means, stds, outputs_for_each_layer
    """
    means, stds, layer_outputs = [], [], []

    # Start with a single input sample of shape (d0,)
    x = np.random.randn(dims[0])

    # Go layer-by-layer
    for i in range(len(dims) - 1):
        W = init_func(dims[i], dims[i+1])
        x = x.dot(W)
        x = activation_func(x)
        means.append(x.mean())
        stds.append(x.std())
        layer_outputs.append(x)

    return means, stds, layer_outputs
```

```python
# ------- PLOTTING HELPER -------
def plot_histograms(layer_outputs, means, stds, title_prefix):
    """
    Plots the histogram of outputs for each layer in a row of subplots.
    """
    num_layers = len(layer_outputs)
    fig, axes = plt.subplots(1, num_layers, figsize=(16, 3))

    if num_layers == 1:
        axes = [axes]  # so we can index axes[0]

    for i, ax in enumerate(axes):
        ax.hist(layer_outputs[i], bins=50, density=True)
        ax.set_title(f"{title_prefix} L{i+1}\nmean={means[i]:.2f}, std={stds[i]:.2f}")
        ax.set_xlim([-1.2, 1.2])  # to compare easily across layers/initializations
    plt.tight_layout()
    plt.show()
```

```python
# Example: 7 layers each of dimension 4096
dims = [4096] * 7

# ---------- 1) Random init + tanh ----------
means, stds, outs = forward_pass(dims, init_random, tanh)
plot_histograms(outs, means, stds, "Random init + tanh")

# ---------- 2a) Xavier + tanh ----------
means, stds, outs = forward_pass(dims, init_xavier, tanh)
plot_histograms(outs, means, stds, "Xavier (Glorot) + tanh")

# ---------- 2b) Xavier + ReLU ----------
means, stds, outs = forward_pass(dims, init_xavier, relu)
plot_histograms(outs, means, stds, "Xavier + ReLU")

# ---------- 3) He + ReLU ----------
means, stds, outs = forward_pass(dims, init_he, relu)
plot_histograms(outs, means, stds, "He (Kaiming) + ReLU")
```

Now let's compute the gradient vector for a given set of parameters:

\begin{equation}
\frac{\partial L}{\partial \boldsymbol\phi} = \begin{bmatrix}\frac{\partial L}{\partial \phi_0} \\\frac{\partial L}{\partial \phi_1} \end{bmatrix}.
\end{equation}

```python
# These came from writing out the expression for the sum of squares loss and taking the
# derivative with respect to phi0 and phi1. It was a lot of hassle to get it right!

def gabor_deriv_phi0(data_x,data_y,phi0, phi1):
    x = 0.06 * phi1 * data_x + phi0
    y = data_y
    cos_component = np.cos(x)
    sin_component = np.sin(x)
    gauss_component = np.exp(-0.5 * x *x / 16)
    deriv = cos_component * gauss_component - sin_component * gauss_component * x / 16
    deriv = 2* deriv * (sin_component * gauss_component - y)
    return np.sum(deriv)

def gabor_deriv_phi1(data_x, data_y,phi0, phi1):
    x = 0.06 * phi1 * data_x + phi0
    y = data_y
    cos_component = np.cos(x)
    sin_component = np.sin(x)
    gauss_component = np.exp(-0.5 * x *x / 16)
    deriv = 0.06 * data_x * cos_component * gauss_component - 0.06 * data_x*sin_component * gauss_component * x / 16
    deriv = 2*deriv * (sin_component * gauss_component - y)
    return np.sum(deriv)

def compute_gradient(data_x, data_y, phi):
    dl_dphi0 = gabor_deriv_phi0(data_x, data_y, phi[0],phi[1])
    dl_dphi1 = gabor_deriv_phi1(data_x, data_y, phi[0],phi[1])
    # Return the gradient
    return np.array([[dl_dphi0],[dl_dphi1]])
```

Now we are ready to find the minimum.  For simplicity, we'll just use regular (non-stochastic) gradient descent with a fixed learning rate.

```python
def gradient_descent_step(phi, data,  model):
  # Step 1:  Compute the gradient
  gradient = compute_gradient(data[0,:],data[1,:], phi)
  # Step 2:  Update the parameters -- note we want to search in the negative (downhill direction)
  alpha = 0.1
  phi = phi - alpha * gradient
  return phi
```
